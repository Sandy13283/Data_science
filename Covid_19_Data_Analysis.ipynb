# Dropping the column

```python
covid=covid.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],axis=1)
```

# Change the format of the Date in the data set

```python
covid['Date']=pd.to_datetime(covid['Date'],format='%Y-%m-%d')
```

```python
covid.head()
```

# Creating new column of the Active cases

= **Confirmed cases-(cured+Deaths)**

```python
covid['Active']=covid['Confirmed']-(covid['Cured']+covid['Deaths'])
```

Fetching bottom 5 rows of the Covid dataset

```python
covid.tail()
```

Creating pivot table with columns with desired colimns

-Confirmed cases

-Deaths

-Active

-Cured

And State names are used as indexed

```python
statewise=pd.pivot_table(covid,values=['Confirmed','Deaths','Active','Cured'],
                         index='State/UnionTerritory',aggfunc=max)
```

```python
statewise
```

# Creating new column

# Recovery Rate

=(Cured*100)/Confirmed
# death Rate
=(Death*100)/Confirmed

```python
statewise['Recovery_rate']=statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['Death_rate']=statewise['Deaths']*100/statewise['Confirmed']
```

Sorting the columns according to the Confirmed cases

```python
statewise=statewise.sort_values(by='Confirmed',ascending=False)
```

# Color coding of the Pivot table

```python
statewise.style.background_gradient(cmap='summer')
```

#Top 10 active cases States

```python
top_10_state=covid.groupby(by='State/UnionTerritory').max()[['Active','Date']].sort_values(by=['Active'],ascending=False).reset_index()
```

```python
top_10_state
```

# Plotting the Top 10 active cases state in a Barplot

```python
fig=plt.figure(figsize=(16,9))
plt.title("Top 10 State with Most Active cases",size=30,color='Purple')
ax=sns.barplot(data=top_10_state.iloc[:10],y="Active",x="State/UnionTerritory",edgecolor="Orange")
plt.xlabel("States")
plt.ylabel("Active Cases")
```

#top state with highest death rates

```python
top_10_Death_state=covid.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
```

```python
plt.figure(figsize=(16,9))
plt.title("Top 10 State with Most deaths",size=30,color='Purple')
sns.barplot(data=top_10_Death_state.iloc[:12],y="Deaths",x="State/UnionTerritory",edgecolor='Green')
```

```python
plt.figure(figsize=(10,6))
plt.title('Top 5 Affected states in india',size=30,color="Green")
sns.lineplot(data=covid[covid['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerala','Tamil Nadu','Uttar Pardesh'])],x='Date',y='Active',hue='State/UnionTerritory')
```

# Vaccine Analysis

```python
vaccine.head(4)
```

```python
vaccine.rename(columns={'Updated On':'Vaccination Date'},inplace=True)
```

```python
vaccine.info()
```

```python
# getting sum of missing values of each columns
vaccine.isnull().sum()
```

```python
Vaccination=vaccine.drop(['Sputnik V (Doses Administered)','AEFI','18-44 Years (Doses Administered)','45-60 Years (Doses Administered)','60+ Years(Individuals Vaccinated)'],axis=1)
```

```python
Vaccination.head()
```

```python
male=Vaccination['Male(Individuals Vaccinated)'].sum()
female=Vaccination['Female(Individuals Vaccinated)'].sum()
px.pie(names=['male','female'],values=[male,female],title='Male and Femation Vaccination')
```

```python
Vaccine=vaccine[vaccine.State!="India"]
```

```python
Vaccine
```

```python
Vaccine.rename(columns={'Total Individuals Vaccinated': "Total"},inplace=True)
Vaccine.head(4)
```

# State with most Vaccination

```python
max_vacc=Vaccine.groupby("State")['Total'].sum().to_frame('Total')
max_vacc=max_vacc.sort_values('Total',ascending=False)[:5]
max_vacc
```

```python
plt.figure(figsize=(10,5))
plt.title("Top 5 Vaccinated states in india",size=30,color='Blue')
sns.barplot(data=max_vacc.iloc[:10],y=max_vacc.Total,x=max_vacc.index,edgecolor="green")
plt.show()
```

# bottom 5 vaccinated states in india

```python
min_vac=Vaccine.groupby("State")['Total'].sum().to_frame('Total')
min_vac=min_vac.sort_values('Total',ascending=True)[:5]
min_vac
```

```python
plt.figure(figsize=(12,7))
plt.title("Bottom 5 Vaccinated States in India")
sns.barplot(data=min_vac.iloc[:10],y=min_vac.Total,x=min_vac.index,linewidth=4,edgecolor='Red')
plt.show()
```
