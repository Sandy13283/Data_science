# **LISTS [ ]**
ordered collection of elements and they are mutable means can be easily changeable.

```python
lis=[1,"Sparta",3.14,1+3j,True]
type(lis[3])
type(lis)
# value assigning
lis[0]=100
# adding an item at last
lis.append(101)
#adding an item at specific position and rest move by 1 index
lis.insert(1,1500)
#removing an last item
lis.pop()
#reversing the elements of list
lis.reverse()
#sorting the list
list1=[1,2,3,4,4,5,56,6,7]
list1.sort()
#adding or concatinating two lists
lis+list1
```

# **Dictionary  { }**
dictionary is the unordered collection of data and it is mutable

```python
dic={'Apple':100,'Litchi':90,'Mangoes':60,}
dic2={"garlic":43,"Potato":40,"Onion":30,"Tomatoe":45,"Carrot":120}
# if we need to extract keys and values seprately from dictionary then
dic.keys()
dic.values()
#changing values inside dictionary
dic["Mangoes"]=65
# adding an another value
dic["Grapes"]=120
dic["Kiwi"]=150
#if we need to update dic in dic2 we can do it
dic.update(dic2)
# To remove an item
dic.pop('garlic')
print(dic)
```

# **Sets { }**
unordered and unindexed collection of elements enclosed with .

Sets are mutable

No duplicates are allowed in it.

```python
s1={1,2,3.14,3.14,"Sparta",True,0,False}
s2={"Sandy",10,20,30,"sparta"}
# adding a single item in sets
s1.add(315)
#adding multiple items
s1.update([10,20,30,40,50,7003,"Badmoshi"])
#remove an element
s1.remove(7003)
s1.remove(False)#zero , 0 is treated as false
# to check intersection or comman
s1.intersection(s2)
# to conclude union
s1.union(s2)
```

# **Creating class with Constructor**



```python
class Employee:
  def __init__(self,name,age,salary,gender):
    self.name=name
    self.age=age
    self.salary=salary
    self.gender=gender

  def employee_details(self):
      print(self.name)
      print(self.age)
      print(self.salary)
      print(self.gender)
E=Employee('Ram',22,85000,'Male')
E.employee_details()
```

# **Matplotlib library**

```python
import numpy as np
from matplotlib import pyplot as plt
x=np.arange(1,11)
y=2*x
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color='Orange',linestyle=':',linewidth=1)
plt.title("X v/s Y")
plt.xlabel("X-axis ")
plt.ylabel("Y-axis ")
plt.show()
```

We can also change line style and color

```python
plt.plot(x,y,color='Orange',linestyle=':',linewidth=2)
plt.show()
```

Showing two graphs in single graphs

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
y3=5*x
plt.plot(x,y1,color='Red',linestyle=':',linewidth=1)
plt.plot(x,y2,color='Yellow',linestyle='-.',linewidth=2)
plt.plot(x,y3,color='Green',linestyle='dashdot',linewidth=3)
plt.title('X and Y ')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

***Plotting two different plot of same thing in different graphs at same-time***

```python
x=np.arange(1,20)
y1=2*x
y2=3*x
y3=5*x
plt.subplot(1,3,3)
plt.plot(x,y1,color='Red',linestyle=':',linewidth=1)

plt.subplot(1,3,2)
plt.plot(x,y2,color='Yellow',linestyle='-.',linewidth=2)

plt.subplot(1,3,1)
plt.plot(x,y3,color='Green',linestyle='dashdot',linewidth=3)
plt.title('X and Y ')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```
